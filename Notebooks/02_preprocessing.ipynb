# 02 Preprocesamiento de datos

En esta etapa se transforma el dataset crudo en un conjunto estructurado y apto para entrenamiento supervisado.

El objetivo principal es garantizar consistencia estadística, compatibilidad matemática con algoritmos de machine learning y minimizar ruido analítico que pueda deteriorar la capacidad predictiva.

El preprocesamiento constituye una fase crítica porque la calidad del modelo depende directamente de la calidad del dato de entrada.

## 1. Importar Librerias

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## 2. Carga del Dataset

```python
df = pd.read_csv('../Data/Raw/Telco-Customer-Churn.csv')
df.head()
```

## 3. Inspección inicial

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Se realiza el cambio de la variable TotalCharges de object a valor numerico

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.isnull().sum()
```

* Al realizar el cambio de la variable TotalCharges a numerica, se identifica que cuenta con registros nulos que deben de ser tratados.

## 4. Tratamiento de dataset

### Eliminar registros nulos

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

* Posterior al tratamiento de la informacion derivado de la eliminacion de los datos nulos, se pudo identificar una modificacion en el numero total de registros contenidos en el dataset, sin embargo al ser menor no se considera que comprometa los resultados del analisis. Estos se pueden considerar que corresponder a clientes recién incorporados sin historial acumulado.

### Eliminar variable costumerID

Derivado del analisis de la informacion, se considera que para los analisis posteriores asi como las predicciones y creacion de modelos de predictivos

```python
df.drop('customerID', axis=1, inplace=True)
df.head()
```

## 5. Codificación de variable objetivo

La variable objetivo en este caso Churn, se cambiara de tipo de dato a binario para poder tratar mas facil la informacion haciendola interpretable por algoritmos de clasificación.

```python
df['Churn'] = df['Churn'].map({
    'No': 0,
    'Yes': 1
})

df['Churn'].value_counts()
```

## 6. Identificar variables categóricas

```python
#Identificar aquellas variables que requieren que se realice codificación para poder realizar un tratamiento matemático.

df.select_dtypes(include='object').columns
```

```python
# Convertir variables categóricas en representación numérica evitando relaciones ordinales artificiales.

df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.shape
```

```python
df_encoded.head()
```

## 7. Separación de variables

```python
# Definir matriz de entrada (X) y variable objetivo (y) para entrenamiento supervisado.

X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

X.shape, y.shape
```

## 8. Train-Test

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)

print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

## 9. Escalado de variables

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled[:5]
```

```python
print('Train:', X_train_scaled.shape)
print('Test:', X_test_scaled.shape)
```

## 10. Exportar datasets procesados

```python
pd.DataFrame(X_train_scaled).to_csv('../data/processed/X_train.csv', index=False)
pd.DataFrame(X_test_scaled).to_csv('../data/processed/X_test.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)
y_test.to_csv('../data/processed/y_test.csv', index=False)
```
