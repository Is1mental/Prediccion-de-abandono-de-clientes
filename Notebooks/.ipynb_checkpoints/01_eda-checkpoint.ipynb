# Exploratory Data Analysis (EDA)

Una empresa de telecomunicaciones enfrenta una alta tasa de cancelación de clientes (churn).
El costo de adquirir nuevos clientes es significativamente mayor que retener a los existentes, por lo que el objetivo es:

*Comprender el comportamiento de los clientes, identificar patrones asociados al churn y detectar variables clave que influyen en la cancelación del servicio, para orientar el modelado predictivo y las decisiones de negocio.*

## 1. Introducción
La retención de clientes es un factor crítico en la industria de telecomunicaciones, ya que adquirir nuevos clientes suele ser más costoso que mantener a los existentes. La pérdida de clientes (churn) impacta directamente en los ingresos y en la estabilidad del negocio.

El objetivo de este proyecto es realizar un análisis exploratorio de datos (EDA) sobre un conjunto de clientes de una empresa de telecomunicaciones, con el fin de:

* Comprender la estructura y calidad de los datos
* Identificar patrones y variables potencialmente relacionadas con el churn
* Preparar el dataset para etapas posteriores de modelado predictivo

## 2. Importación de librerías y configuración

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_context("notebook")

pd.set_option("display.max_columns", None)
```

```python
# Carga del Dataset
df = pd.read_csv("../data/raw/Telco-Customer-Churn.csv")
# Verificacion de datos
df.head()
df.shape
df.info()
```

### Revision de datos
En una primera inspección se identificó que:

* El dataset cuenta con 7,043 registros completos, sin valores nulos explícitos.
* La mayoría de las variables son de tipo categórico, lo cual es común en datasets de comportamiento de clientes.
* Existen 3 variables numéricas (SeniorCitizen, tenure, MonthlyCharges).
* La variable TotalCharges se encuentra almacenada como tipo object, a pesar de representar un valor numérico.

Esta revisión inicial permite validar que el dataset tiene una estructura sólida, pero también revela la necesidad de realizar ajustes antes de un análisis más profundo.

# 3. Limpieza de datos

```python
#Conversion correcta de la variable TotalCharges de object a numerico
# Los valores no convertibles se transforman en NaN

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Revisión de valores nulos por columna
df.isna().sum()
```

```python
# Comparación visual de la antigüedad
sns.boxplot(
    data=df,
    x=df["TotalCharges"].isna(),
    y="tenure"
)

plt.title("Antigüedad (tenure) según TotalCharges nulo o no")
plt.xlabel("TotalCharges es nulo")
plt.ylabel("Antigüedad (meses)")
plt.show()
```

Los valores nulos en la variable `TotalCharges` corresponden casi exclusivamente a clientes con antigüedad cercana a 0 meses, lo cual es consistente con la lógica de negocio: clientes recién incorporados aún no han generado cargos acumulados.
Por lo tanto, la ausencia de valores no es aleatoria y refleja una condición operativa del negocio.

```python
# Eliminación de filas con valores nulos representan un porcentaje pequeño del total
df = df.dropna()
# Verificación de dimensiones después de la limpieza
df.shape
```

## 4. Análisis de la variable objetivo: Churn

```python
# Distribución porcentual de la variable objetivo
df["Churn"].value_counts(normalize=True) * 100
```

```python
# Visualización de la distribución de churn
sns.countplot(data=df, x="Churn")
plt.title("Distribución de Churn")
plt.xlabel("Churn")
plt.ylabel("Número de clientes")
plt.show()
```

En el grafico se puede apreciar que aproximadamente una cuarta parte de los clientes cancelan el servicio

### Análisis de Churn vs Antigüedad (Tenure) 

```python
# Distribución de la antigüedad segmentada por churn
sns.histplot(
    data=df,
    x="tenure",
    hue="Churn",
    bins=30,
    kde=True
)

plt.title("Distribución de Antigüedad por Churn")
plt.xlabel("Antigüedad (meses)")
plt.ylabel("Frecuencia")
plt.show()
```

De presente analisis se puede determinar que los clientes con baja antigüedad presentan una mayor probabilidad de churn.

### Análisis de Churn por Tipo de Contrato

```python
# Comparación de churn según tipo de contrato
sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.title("Churn por Tipo de Contrato")
plt.xlabel("Tipo de contrato")
plt.ylabel("Número de clientes")
plt.xticks(rotation=15)
plt.show()
```

Haciendo el analisis por tipo de contrato, se puede determinar que los contratos mensuales concentran la mayor tasa de cancelación.

### Churn vs Cargos Mensuales

```python
# Comparación de cargos mensuales entre clientes que cancelan y los que no
sns.boxplot(
    data=df,
    x="Churn",
    y="MonthlyCharges"
)

plt.title("Cargos Mensuales vs Churn")
plt.xlabel("Churn")
plt.ylabel("Cargos Mensuales")
plt.show()
```

Se puede identificar que se tienen cargos mensuales elevados incrementan el riesgo de churn, especialmente en contratos flexibles.

## Correlación entre variables

```python
numeric_df = df.select_dtypes(include=["int64","float64"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlación entre variables numéricas")
plt.show()
```
